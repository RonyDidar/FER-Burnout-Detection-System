```python
### ========================================================================================================================================
## Name:        Saputra Rizky Johan (사푸트라)
## Institution: Seoul National University, South Korea
## Purpose:     Undergraduate Thesis
## Department:  Computer Science and Engineering
## College:     Engineering
## Advisor:     Prof. Lee Changun (이창건 교수)
## Title:       The Application of Facial Emotion Recognition (FER) in the Detection and Measurement of Burnout and Prolonged Stress Levels
## Module:      fer_display_visualizer.ipynb (A notebook file to append the FER display images into a single figure for thesis purposes)
### ========================================================================================================================================

# =============================================================================================
# SETUP
# =============================================================================================
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# =============================================================================================
# IMPLEMENTATION
# =============================================================================================
# Initialize the list of image files for the figure
files = ['fer_display1.png', 'fer_display2.png', 'fer_display3.png']

# Declare the title for each subfigure
titles = [
    r'User001: $s_{system}$ = 0.400, $s_{survey}$ = 0.600',
    r'User003: $s_{system}$ = 0.717, $s_{survey}$ = 0.733',
    r'User005: $s_{system}$ = 0.706, $s_{survey}$ = 0.867'
]

# Initialize the figure and its axes
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot each of the specified images
for ax, file, title in zip(axes, files, titles):
    img = mpimg.imread(f'figures/{file}')
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off')  # Hide axes

# Declare the main title for the figure
fig.suptitle(
    'Figure 1. FER System Display: Real-Time Emotion Classifier and Burnout Estimator',
    fontsize=14,
    fontweight='bold',
)

# Adjust the layout for spacing purposes
plt.tight_layout()

# Save the figure for thesis documentation (Adjust if necessary)
plt.savefig('fer_display_combined.png', dpi=300, bbox_inches='tight')

# Load the figure on the notebook
plt.show()

### ===================================================================================================================================
## END: Add implementations if necessary
### ===================================================================================================================================
```
