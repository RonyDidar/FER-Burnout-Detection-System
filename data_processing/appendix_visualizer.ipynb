```python
# =============================================================================================
# SETUP
# =============================================================================================
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# =============================================================================================
# IMPLEMENTATION (MAIN)
# =============================================================================================
# Initialize the list of image files for the figure
files = ['main_snippet1.png', 'main_snippet2.png', 'main_snippet3.png']

# Declare the title for each subfigure
titles = [
    "Code Snippet 1",
    "Code Snippet 2",
    "Code Snippet 3"
]

# Initialize the figure and its axes
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot each of the specified images
for ax, file, title in zip(axes, files, titles):
    img = mpimg.imread(f'code_snippets/{file}')
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off')  # Hide axes

# Declare the main title for the figure
fig.suptitle(
    'Figure 5. The Code Snippet for main.py',
    fontsize=14,
    fontweight='bold',
)

# Adjust the layout for spacing purposes
plt.tight_layout()

# Save the figure for thesis documentation (Adjust if necessary)
plt.savefig('main_combined.png', dpi=300, bbox_inches='tight')

# Load the figure on the notebook
plt.show()
```
